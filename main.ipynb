```python
genai.configure(api_key="AIzaSyBtYB7Da8VNN8paGnsA2fSNx69UqYfTXao")
genre = input("Enter the genre of the movie: ")
characters = input("Enter the characters of the movie: ")
scene_description = input("Enter the scene description: ")

# Create the model
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 40,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
  model_name="gemini-2.0-flash-exp",
  generation_config=generation_config,
)

# Generate content based on user inputs
prompt = f"Generate a movie script: Genre: {genre}, Characters: {characters}, Scene Description: {scene_description}"
response = model.generate_content(prompt, stream=True)

# Print the generated content
for chunk in response:
    print(chunk.text, end="") 
```
